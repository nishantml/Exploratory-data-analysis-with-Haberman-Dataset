```python
#Histogram
import numpy as np
survival_less_than_5_years = ds.loc[ds['Surv_status'] == 2];
survival_more_than_5_years = ds.loc[ds['Surv_status'] == 1];
plt.plot(survival_less_than_5_years['age'],np.zeros_like(survival_less_than_5_years['age']),'o')
plt.plot(survival_more_than_5_years['age'],np.zeros_like(survival_more_than_5_years['age']),'v')

plt.show()
```

```python
sns.FacetGrid(ds, hue="Surv_status", size=5) \
   .map(sns.distplot, "age") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(ds, hue="Surv_status", size=5) \
   .map(sns.distplot, "Operation_Year") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(ds, hue="Surv_status", size=5) \
   .map(sns.distplot, "axil_nodes") \
   .add_legend();
plt.show();
```

```python
#observation 
#here we see that axil_node is less than 5 then there is more chance of survival
survival = ds[(ds['Surv_status'] == 1) & (ds['axil_nodes']<5)]
survival.count()
```

```python
counts,bin_edges = np.histogram(survival_more_than_5_years['axil_nodes'],bins=10,density = True)

pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:],cdf);

counts,bin_edges = np.histogram(survival_less_than_5_years['axil_nodes'],bins=10,density = True)

pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:],cdf);


plt.show();
```

```python
#Observations
#here we see that 82% of survival rate if axil_node is less than 8
#here we also see that if the axil_nodes between 10 to 28 the death rate is lies between 70%  to 100%

```

```python
#mean ,Variance and std-dev
print('Mean');
print(np.mean(survival_less_than_5_years['axil_nodes']))
print(np.mean(survival_more_than_5_years['axil_nodes']))

print('\nstd');
print(np.std(survival_less_than_5_years['axil_nodes']))
print(np.std(survival_more_than_5_years['axil_nodes']))
```

```python
print('Median')
print(np.median(survival_less_than_5_years['axil_nodes']))
print(np.median(survival_more_than_5_years['axil_nodes']))

print('\n Quantiles');
print(np.percentile(survival_more_than_5_years['axil_nodes'],np.arange(0,100,25)))
print(np.percentile(survival_more_than_5_years['Operation_Year'],np.arange(0,100,25)))
print(np.percentile(survival_more_than_5_years['age'],np.arange(0,100,25)))

print('\n 90th Percentile')
print(np.percentile(survival_more_than_5_years['axil_nodes'],90))
print(np.percentile(survival_more_than_5_years['Operation_Year'],90))
print(np.percentile(survival_more_than_5_years['age'],90))

```
